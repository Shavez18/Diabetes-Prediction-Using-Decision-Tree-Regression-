```python
## Load Dataset Using Scikit_Learn Library.

from sklearn.datasets import load_diabetes
dataset= load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR']) # Description of the dataset
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
df_diabetes.head()
```

```python
## Independent and dependent features
X=df_diabetes
y=dataset['target']
```

```python
## Applying Train Test Split.

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head() # Training set features
```

```python
## correlation
X_train.corr()
```

```python
## Visualize Correaltion Heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('Correlation Matrix')
sns.heatmap(X_train.corr(),annot=True,fmt='.2f',cmap='coolwarm') 
plt.show()
```

```python
# Create Decision Tree Regressor Object.
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor() 
regressor.fit(X_train,y_train)
```

```python
## Hyperparameter Tunning.

param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore') 
grid.fit(X_train,y_train) 
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
print(y_pred)
```

```python
# Importing evaluation metrics.
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print('R2_Score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
##Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel,filled=True)
```
